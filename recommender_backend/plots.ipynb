```python
import matplotlib.pyplot as plt
import pandas as pd

# Data extracted directly from your output for the Women's evaluation
ocs_scores_women = {
    'casual_weekend': 0.8090,
    'sporty': 0.8039,
    'party': 0.7903,
    'job_interview': 0.8090
}

# Convert to a DataFrame for easy plotting and sorting
df_women = pd.DataFrame(list(ocs_scores_women.items()), columns=['Context', 'OCS'])
df_women = df_women.sort_values(by='OCS', ascending=False)

# --- Plot Generation ---
plt.figure(figsize=(10, 6))
bars = plt.bar(df_women['Context'], df_women['OCS'], color='#C2185B', zorder=2) # Deep pink color

# Set Y-axis limits to clearly show the small differences near 0.80
plt.ylim(0.78, 0.82) 
plt.axhline(df_women['OCS'].mean(), color='gray', linestyle='--', linewidth=1.5, label=f'Mean OCS: {df_women["OCS"].mean():.3f}')


plt.title('Outfit Completeness Score (OCS) by Context (Women)', fontsize=16, fontweight='bold')
plt.ylabel('Mean Cosine Similarity', fontsize=12)
plt.xlabel('Outfit Context', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6, zorder=1)
plt.legend()

# Add labels to bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.0005,
             f'{yval:.4f}',
             ha='center', va='bottom',
             size=11, fontweight='bold')
             
plt.tight_layout()
plot_filename = 'final_ocs_plot_women.png'
plt.savefig(plot_filename)
print(f"Plot saved as {plot_filename}")
```
